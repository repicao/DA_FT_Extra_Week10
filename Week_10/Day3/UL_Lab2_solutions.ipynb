
# Unsupervised Learning Lab: Evaluation Metrics - Solutions

## Exercise 1: Calculate Silhouette Score
```python
import pandas as pd
from sklearn.metrics import silhouette_score

# Load the dataset
df = pd.read_csv('customer_data.csv')

# Assuming the clustering results from the previous lab are stored in the dataframe
# df['KMeans_Cluster'], df['Hierarchical_Cluster'], df['DBSCAN_Cluster']

# Calculate Silhouette Score for K-Means
kmeans_silhouette = silhouette_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['KMeans_Cluster'])
print(f'Silhouette Score for K-Means: {kmeans_silhouette}')

# Calculate Silhouette Score for Hierarchical Clustering
hierarchical_silhouette = silhouette_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['Hierarchical_Cluster'])
print(f'Silhouette Score for Hierarchical Clustering: {hierarchical_silhouette}')

# Calculate Silhouette Score for DBSCAN
dbscan_silhouette = silhouette_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['DBSCAN_Cluster'])
print(f'Silhouette Score for DBSCAN: {dbscan_silhouette}')
```

## Exercise 2: Calculate Davies-Bouldin Index
```python
from sklearn.metrics import davies_bouldin_score

# Calculate Davies-Bouldin Index for K-Means
kmeans_db_index = davies_bouldin_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['KMeans_Cluster'])
print(f'Davies-Bouldin Index for K-Means: {kmeans_db_index}')

# Calculate Davies-Bouldin Index for Hierarchical Clustering
hierarchical_db_index = davies_bouldin_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['Hierarchical_Cluster'])
print(f'Davies-Bouldin Index for Hierarchical Clustering: {hierarchical_db_index}')

# Calculate Davies-Bouldin Index for DBSCAN
dbscan_db_index = davies_bouldin_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['DBSCAN_Cluster'])
print(f'Davies-Bouldin Index for DBSCAN: {dbscan_db_index}')
```

## Exercise 3: Calculate Calinski-Harabasz Index
```python
from sklearn.metrics import calinski_harabasz_score

# Calculate Calinski-Harabasz Index for K-Means
kmeans_ch_index = calinski_harabasz_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['KMeans_Cluster'])
print(f'Calinski-Harabasz Index for K-Means: {kmeans_ch_index}')

# Calculate Calinski-Harabasz Index for Hierarchical Clustering
hierarchical_ch_index = calinski_harabasz_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['Hierarchical_Cluster'])
print(f'Calinski-Harabasz Index for Hierarchical Clustering: {hierarchical_ch_index}')

# Calculate Calinski-Harabasz Index for DBSCAN
dbscan_ch_index = calinski_harabasz_score(df[['Age', 'AnnualIncome', 'SpendingScore']], df['DBSCAN_Cluster'])
print(f'Calinski-Harabasz Index for DBSCAN: {dbscan_ch_index}')
```
